````python
#Storing collection in a list
Collection_list=zomato_df.Collections.dropna().str.split(', ')
#Storing all Collections in dict
Collections_dict={}
for collection in Collection_list:
    for col_name in collection:
        if (col_name in Collections_dict):
            Collections_dict[col_name]+=1
        else:
            Collections_dict[col_name]=1
````

````python
#converting the collection dict in the new dataframe
Collection_df=pd.DataFrame([Collections_dict]).transpose().reset_index().rename(columns={'index':'Tagging',0:'No_of_Restaurants'})
````

````python
#Storing cuisines in a list
Cuisine_list=zomato_df.Cuisines.dropna().str.split(', ')
#Storing all cuisines in dict
Cuisine_dict={}
for cuisine in Cuisine_list:
    for col_name in cuisine:
        if (col_name in Cuisine_dict):
            Cuisine_dict[col_name]+=1
        else:
            Cuisine_dict[col_name]=1
````

````python
Cuisine_df=pd.DataFrame([Cuisine_dict]).transpose().reset_index().rename(columns={'index':'Food_type',0:'No_of_Restaurants'})
````

Wrangling on reviews dataset

````python
reviews_df['Rating'].unique()
````

````python
# Converting rating variable in float by converting the string to null
reviews_df.loc[reviews_df['Rating']=='Like']=np.nan
reviews_df['Rating']= reviews_df['Rating'].astype('float')

#Fill null values with mean
reviews_df['Rating'].fillna(3.5, inplace=True)
````

````python
# Custom function to extract reviews and followers from metadata
def extract_reviews_followers(metadata):
    if isinstance(metadata, str):
        parts = metadata.split(',')
        if len(parts) == 2:  # Check if both reviews and followers are present
            reviews = parts[0].strip().split()[0]  # Extract numeric value for reviews
            followers = parts[1].strip().split()[0]  # Extract numeric value for followers
            return pd.Series([reviews, followers])
    return pd.Series([0, 0])  # Handle None or non-string values

# Apply the custom function to extract reviews and followers
split_data = reviews_df['Metadata'].apply(extract_reviews_followers)

# Assign the extracted data to new columns
reviews_df['Reviewers_Reviews'] = pd.to_numeric(split_data[0], errors='coerce')
reviews_df['Reviewers_Followers'] = pd.to_numeric(split_data[1], errors='coerce')

# Display the resulting DataFrame
reviews_df
````

````python
# Adding year and month columns for clear insights
reviews_df['Year'] = pd.DatetimeIndex(reviews_df['Time']).year
reviews_df['Month'] = pd.DatetimeIndex(reviews_df['Time']).month
````

````python
#review counts
reviewer_list = reviews_df.groupby('Reviewer').apply(lambda x: x['Reviewer'].count()).reset_index(name='Review_Count')
````

````python
#Top 10 reviewers
top_reviewers=reviewer_list.sort_values('Review_Count', ascending=False)[:10]
top_reviewers
````

````python
#Top 10 rated restaurants
avg_rating = reviews_df.groupby('Restaurant').agg({'Rating':'mean','Reviewer': 'count'}).reset_index().rename(columns = {'Reviewer': 'Total_Review'})
top_rated_restaurant=avg_rating.sort_values('Rating', ascending=False)[:10]
top_rated_restaurant
````

````python
#combining both the dataframe for effective insights
zomato_df = zomato_df.rename(columns = {'Name':'Restaurant'})
df = zomato_df.merge(reviews_df, on = 'Restaurant')
df.shape
````

````python
df.head()
````

````python
# Handling Outliers & Outlier treatments
# To separate the symmetric distributed features and skew symmetric distributed features
symmetric_feature=[]
non_symmetric_feature=[]
for i in df.describe().columns:
  if abs(df[i].mean()-df[i].median())<0.2:
    symmetric_feature.append(i)
  else:
    non_symmetric_feature.append(i)

# Getting Symmetric Distributed Features
print("Symmetric Distributed Features : -",symmetric_feature)

# Getting Skew Symmetric Distributed Features
print("Skew Symmetric Distributed Features : -",non_symmetric_feature)
````

````python
# For Skew Symmetric features defining upper and lower boundry
#Outer Fence
def outlier_treatment_skew(df,feature):
  IQR= df[feature].quantile(0.75)- df[feature].quantile(0.25)
  lower_bridge =df[feature].quantile(0.25)- 1.5*IQR
  upper_bridge =df[feature].quantile(0.75)+ 1.5*IQR
  # print(f'upper : {upper_bridge} lower : {lower_bridge}')
  return upper_bridge,lower_bridge
````

````python
# Restricting the data to lower and upper boundary for cost in hotel dataset
#lower limit capping
zomato_df.loc[zomato_df['Cost']<= outlier_treatment_skew(df=zomato_df, feature='Cost')[1], 'Cost']=outlier_treatment_skew(df=zomato_df,feature='Cost')[1]

#upper limit capping
zomato_df.loc[zomato_df['Cost']>= outlier_treatment_skew(df=zomato_df, feature='Cost')[0], 'Cost']=outlier_treatment_skew(df=zomato_df,feature='Cost')[0]
````

##### Which text vectorization technique have you used and why?

Tf-idf Vectorization is a powerful technique used to convert text data into numerical representations that machine learning models can understand. It operates by assigning weights to words based on their frequency in a document and their rarity across a collection of documents.

In essence, tf-idf evaluates the importance of a word within a document by considering both how often it appears in that document (term frequency) and how uncommon it is across the entire document collection (inverse document frequency). This process helps in identifying words that are unique and informative to a particular document.

The formula for tf-idf involves multiplying the term frequency (tf) by the inverse document frequency (idf) for each word. This calculated weight provides a meaningful representation of the word's significance within the context of the document and the overall collection.

One of the key benefits of tf-idf is its ability to reduce the dimensionality of text data while highlighting important words that carry more meaning. This feature weighting technique enhances the performance of machine learning models, especially in tasks like text classification and information retrieval, where distinguishing between relevant and irrelevant words is crucial.

Text vectorization, including tf-idf, plays a vital role in natural language processing tasks by transforming textual data into a format that machine learning algorithms can process effectively. Other text vectorization methods, such as bag-of-words (BoW), word2vec, and doc2vec, offer alternative approaches to encoding text data but tf-idf stands out for its ability to capture the significance of words based on their frequency and rarity across documents.

It is derived by comparing the number of observations in the majority class (Nm) against those in the minority class (Ns).
